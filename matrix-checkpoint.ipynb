```python
import numpy as np
size = 85

M = np.ones((size, size)) * 2 + np.diag(np.arange(1, 170, 2))
print(f'\t\t M:\n{M}')
```

```python
N = np.ones((size, size)) * 4
N[::2, 2::3] = 5
N[1::2, 1::2] = 7
print(f'\t N:\n{N}')
```

```python
mu = 2
sigma = 4
p = 1 + np.random.normal(mu, np.sqrt(sigma), size=size)
print(f'p: {p}')
```

```python
qT = np.multiply(N[24, :size],N[71, :size])
q = qT.T
print(f'q\u1D40: \n{qT}')
print(f'\n q: \n{q}')
```

```python
# f 
n_p = N * p 
sumq = np.sum(n_p[np.arange(85) != 37])
print(f'f: \n{q * sumq}')
```

```python
g = np.sum((p * q)[(np.arange(85) > 11) & (np.arange(85) < 77) & (np.arange(85) != np.diag(M))])
print(f'g: \n{g}')
```
